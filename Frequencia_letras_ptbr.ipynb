```python
with open("lotr2.txt","r", encoding="utf-8") as entrada:
    texto = entrada.read()
print(texto[:2000])
print(len(texto))
```

```python
tamanho_bloco = len(texto)//1_000_000  + 1
tamanho = len(texto) // tamanho_bloco
partes = [texto[i:i+tamanho] for i in range(0, len(texto), tamanho)]
```

```python
import re
import unicodedata

def limpar_texto(texto):
    texto = unicodedata.normalize('NFD', texto)
    texto = texto.encode('ascii', 'ignore').decode('utf-8')    
    texto = texto.replace('-', ' ')
    texto = re.sub(r'[^a-zA-Z\s]', '', texto)
    texto = texto.replace('\n', ' ')
    texto = texto.lower()
    return texto

resultado = set()
for parte in partes:
    aux = limpar_texto(parte)
    aux = aux.split()
    print(len(aux))
    aux = set(aux)    
    print(len(aux))
    resultado.update(aux)
    
# print(parte[0])   
valorresultado = list(resultado)
print(len(valorresultado))
print(valorresultado)
```

```python
import string

def criar_matriz_vazia():
    letras = list(string.ascii_lowercase)
    matriz = [[0 for _ in letras] for _ in letras]
    return matriz

def atualizar_matriz(matriz, palavra):
    letras = list(string.ascii_lowercase)
    pares = [palavra[i:i+2] for i in range(len(palavra) - 1)]

    for par in pares:
        if len(par) == 2:
            i, j = letras.index(par[0]), letras.index(par[1])
            matriz[i][j] += 1

# Exemplo de uso
matriz2 = criar_matriz_vazia()
for palavra in palavras:
    atualizar_matriz(matriz2, palavra)

# Imprimir matriz2
for linha in matriz2:
    print(linha)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plotar_heatmap(matriz):
    letras = list(string.ascii_lowercase)
    plt.figure(figsize=(12, 8))
    cmap = sns.color_palette("Blues", as_cmap=True)
    annot = [[f"{letras[i]}{letras[j]}" for j in range(len(letras))] for i in range(len(letras))]
    sns.heatmap(matriz, xticklabels=letras, yticklabels=letras, cmap=cmap, annot=annot, fmt='', vmin=0.01)
    plt.title("Heatmap de Pares de Letras")
    plt.xlabel("Letra Seguinte")
    plt.ylabel("Letra Atual")
    plt.show()
    
plotar_heatmap(matriz2)
```

```python
def plotar_heatmap_valores(matriz):
    letras = list(string.ascii_lowercase)
    plt.figure(figsize=(14, 10))  # Aumentar o tamanho da figura
    cmap = sns.light_palette("red", as_cmap=True)  # Tons de vermelho de branco para vermelho
    annot = [[f"{valor:.3f}" for valor in linha] for linha in matriz]
    sns.heatmap(
        matriz,
        xticklabels=letras,
        yticklabels=letras,
        cmap=cmap,
        annot=annot,
        fmt='',
        vmin=0,
        linewidths=0.5,  # Largura da borda
        linecolor='white',  # Cor da borda
        annot_kws={"size": 8}  # Reduzir o tamanho da fonte das anotações
    )
    
    plt.title("Heatmap de Frequência de Pares de Letras")
    plt.xlabel("Letra Seguinte")
    plt.ylabel("Letra Atual")
    plt.show()
```

```python
def gerar_matriz_frequencia(matriz):
    total_pares = sum(sum(linha) for linha in matriz)
    if total_pares == 0:
        return matriz  # Evitar divisão por zero

    matriz_frequencia = [[round((valor / total_pares) * 100, 2) for valor in linha] for linha in matriz]
    return matriz_frequencia
matriz_frequencia = gerar_matriz_frequencia(matriz2)
plotar_heatmap_valores(matriz_frequencia)
for linha in matriz_frequencia:
    print(linha)
```
